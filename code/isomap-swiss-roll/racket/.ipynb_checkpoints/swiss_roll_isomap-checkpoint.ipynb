```python
import numpy as np
import matplotlib.pyplot as plt

N =400
mu, sigma = 0, 1 # mean and standard deviation
s1 = np.random.normal(mu, sigma, N)
s11 = np.random.normal(mu, sigma, N)
s2 = np.random.normal(mu, sigma, N)
s22 = np.random.normal(mu, sigma, N)
s3 = np.random.normal(mu, sigma, N)
s33 = np.random.normal(mu, sigma, N)
s4 = np.random.normal(mu, sigma, N)
s44 = np.random.normal(mu, sigma, N)
```

```python
#centers = [7.5 7.5; 7.5 12.5; 12.5 7.5; 12.5 12.5]
s1x = s1+7.5
s1y = s11+7.5
s2x = s2+7.5
s2y = s22+12.5
s3x = s3+12.5
s3y = s33+7.5
s4x = s4+12.5
s4y = s44+12.5
```

```python

plt.plot(s1x, s1y, 'o', color='red');
plt.plot(s2x, s2y, 'o', color='yellow');
plt.plot(s3x, s3y, 'o', color='blue');
plt.plot(s4x, s4y, 'o', color='green');
```

```python
xcos1x = s1x * np.cos(s1x)
xsin1x = s1x * np.sin(s1x)
xcos2x = s2x * np.cos(s2x)
xsin2x = s2x * np.sin(s2x)
xcos3x = s3x * np.cos(s3x)
xsin3x = s3x * np.sin(s3x)
xcos4x = s1x * np.cos(s4x)
xsin4x = s1x * np.sin(s4x)
```

```python
ax = plt.axes(projection='3d')

ax.scatter3D(xcos1x, s1y, xsin1x)
ax.scatter3D(xcos2x, s2y, xsin2x)
ax.scatter3D(xcos3x, s3y, xsin3x)
ax.scatter3D(xcos4x, s4y, xsin4x)
```

```python
from math import sqrt, pow

class ThreeDPoint():
    def __init__(self, x=0, y=0, z =0):
        self.x = x
        self.y = y
        self.z = z
    def dist(self, P):
        return np.sqrt(np.power(self.x - P.x, 2) + np.power(self.y - P.y, 2)+ np.power(self.z - P.z, 2))
    def __str__ (self):
        strings = ("Point(x:%.2f, y:%.2f, z:%.2f)" % (self.x,self.y,self.z))
        return strings
```

```python

point_list = []
for i in range (0, N):
    point_list.append( ThreeDPoint(xcos1x[i], s1y[i], xsin1x[i]))
for i in range (0, N):
    point_list.append(ThreeDPoint(xcos2x[i], s2y[i], xsin2x[i]))
for i in range (0, N):
    point_list.append(ThreeDPoint(xcos3x[i], s3y[i], xsin3x[i]))
for i in range (0, N):
    point_list.append( ThreeDPoint(xcos4x[i], s4y[i], xsin4x[i]))
print(point_list[1])
```

```python
distance =np.zeros((4*N,4*N))
for i in range(0,4*N):
    current_point = point_list[i]
 
    for j in range (0,4*N):
        distance_current = current_point.dist(point_list[j])
  
        distance[i][j] =  distance_current
```

```python
print(distance)
```
